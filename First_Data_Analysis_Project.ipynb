This project analyzes the sales data using Python. It includes data cleaning and visualization.The project will provide a comprehensive view of the electronic market, how sales changes over time, highest sales by City and the relationship between age and salary.


```python
# importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading The Dataset

```python
dataframe = pd.read_csv('/content/Sales_dataset.csv.zip')
print(dataframe.head())
```

Loading The DataFrame Information

```python
print(dataframe.head())
print(dataframe.info())
```

Removing Duplicates

```python
print(dataframe.drop_duplicates())
```

Checking for missing values

```python
print(dataframe.dropna())
```

Visualization

```python
# Line plot

plt.figure()
plt.plot(dataframe["date"], dataframe["sales"])
plt.title("sales over time")
plt.xlabel("date")
plt.ylabel("sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sales Distribution (Histogram)

plt.figure()
plt.hist(dataframe["sales"], bins=10)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# Sales by City (Bar chart)

sales_by_city = dataframe.groupby("city")["sales"].sum()

plt.figure()
plt.bar(sales_by_city.index, sales_by_city.values)
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Sales")
plt.show()
```

```python
# Age vs Salary (Scatter plot)

plt.figure()
plt.scatter(dataframe["age"], dataframe["salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
